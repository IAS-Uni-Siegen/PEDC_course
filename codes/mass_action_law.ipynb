```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.617e-5  # eV/K
T = 300  # Temperature in Kelvin
Eg = 1.12  # eV, band gap for silicon
Nc = 2.8e19  # Effective density of states in conduction band (cm^-3)
Nv = 1.04e19  # Effective density of states in valence band (cm^-3)

# Intrinsic carrier concentration
ni = np.sqrt(Nc * Nv) * np.exp(-Eg / (2 * k * T))

# Doping range
Nd_values = np.logspace(14, 18, 100)  # donor concentrations from 1e14 to 1e18
n = Nd_values
p = ni**2 / n

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(Nd_values, n, label='Electron concentration (n)', color='blue')
plt.plot(Nd_values, p, label='Hole concentration (p)', color='red')
plt.axhline(ni, color='gray', linestyle='--', label=f'Intrinsic carrier conc. (n_i ≈ {ni:.1e})')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Donor Doping Concentration $N_D$ (cm$^{-3}$)')
plt.ylabel('Carrier Concentration (cm$^{-3}$)')
plt.title('Mass Action Law: $n \\cdot p = n_i^2$ at T = 300 K')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
